```python
from sklearn.preprocessing import OneHotEncoder
stroke = pd.concat([stroke, pd.get_dummies(stroke['Physical Activity'], prefix='Physical Activity', drop_first=True)], axis=1)
stroke = pd.concat([stroke, pd.get_dummies(stroke['Work Type'], prefix='Work Type', drop_first=True)], axis=1)
stroke = pd.concat([stroke, pd.get_dummies(stroke['Alcohol Intake'], prefix='Alcohol Intake', drop_first=True)], axis=1)
stroke.drop(['Physical Activity','Work Type','Alcohol Intake'], axis=1, inplace=True)
stroke.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3,random_state=123)
pca.fit(stroke[['Age','Average Glucose Level','Stress Levels']])
princ_comp=pca.transform(stroke[['Age','Average Glucose Level','Stress Levels']])
```

```python
pca.explained_variance_ratio_.round(3)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1, random_state=123)
pca.fit(stroke[['Age','Average Glucose Level','Stress Levels']])
stroke['dimension'] = pca.transform(stroke.loc[:, ('Age','Average Glucose Level','Stress Levels')]).flatten()
stroke.drop(['Age','Average Glucose Level','Stress Levels'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = stroke.drop(["Body Mass Index (BMI)"], axis=1)
y = stroke["Body Mass Index (BMI)"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=123)
```

```python
print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in whole dataset: {len(X_train)}')
print(f'Total # of sample in whole dataset: {len(X_test)}')
```

```python
from sklearn.preprocessing import StandardScaler

numerical_features = ['Stroke History', 'Heart Disease', 'dimension']
scaler = StandardScaler()
scaler.fit(X_train[numerical_features])
X_train[numerical_features] = scaler.transform(X_train.loc[:, numerical_features])
X_train[numerical_features].head()
```

```python
models = pd.DataFrame(index=['train_mse', 'test_mse'],
                      columns=['KNN', 'RandomForest', 'Boosting'])
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor


RF = RandomForestRegressor(n_estimators=50, max_depth=16, random_state=55, n_jobs=-1)
RF.fit(X_train, y_train)

models.loc['train_mse','RandomForest'] = mean_squared_error(y_pred=RF.predict(X_train), y_true=y_train)
```

```python
from sklearn.ensemble import AdaBoostRegressor
boosting = AdaBoostRegressor(n_estimators=50, learning_rate=0.05, random_state=55)
boosting.fit(X_train, y_train)
models.loc['train_mse','Boosting'] = mean_squared_error(y_pred=boosting.predict(X_train), y_true=y_train)
```

```python
mse = pd.DataFrame(columns=['train', 'test'], index=['KNN','RF','Boosting'])
model_dict = {'KNN': knn, 'RF': RF, 'Boosting': boosting}
for name, model in model_dict.items():
    mse.loc[name, 'train'] = mean_squared_error(y_true=y_train, y_pred=model.predict(X_train))/1e3
    mse.loc[name, 'test'] = mean_squared_error(y_true=y_test, y_pred=model.predict(X_test))/1e3
```

```python
fig, ax = plt.subplots()
mse.sort_values(by='test', ascending=False).plot(kind='barh', ax=ax, zorder=3)
ax.grid(zorder=0)
```

```python
prediksi = X_test.iloc[:1].copy()
pred_dict = {'y_true':y_test[:1]}
for name, model in model_dict.items():
    pred_dict['prediksi_'+name] = model.predict(prediksi).round(1)

pd.DataFrame(pred_dict)
```
