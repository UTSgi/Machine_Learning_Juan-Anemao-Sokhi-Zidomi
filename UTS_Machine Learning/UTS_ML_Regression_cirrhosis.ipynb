Prompt Chatgpt:buatkan program untuk melakukan visualisasi hasil prediksi dari regresi

Fungsi dari program dibawah :


1.   Melakukan import library numpy as pengolahan array,pandas untuk melakukan pengolahan dataframe ,library LinearRegression untuk pemodelan ,library mean_squared_error untuk perhitungan MSE,library r2_score untuk perhitungan r2_score dan library matplotlib untuk visualisasi
2.   Membuat dataframe dan melakukan seleksi feature dan target
3.   Melakukan pemisahan data train dan data test
dengan menggunakan library train_test_split
4.   Melakukan pemodelan dengan menggunakan Linear Regression
5.   Melakukan training model
6.   Melakukan prediksi
7.   Menghitung MSE dan R2_score kemudian menampilkannya


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Buat array untuk fitur (features) dan target variable
# Gantilah 'data_bilirubin' dan 'data_cholesterol' dengan data yang sesuai
data_bilirubin = np.array([1.2, 2.3, 3.1, 4.0, 2.8, 1.5, 3.7, 2.2, 1.8, 3.4])
data_cholesterol = np.array([150, 180, 200, 220, 160, 140, 210, 185, 170, 190])
data_stage = np.array([3, 4, 2, 1, 3, 4, 2, 3, 2, 1])

# Gabungkan array menjadi satu array 2D
features = np.column_stack((data_bilirubin, data_cholesterol))
target = data_stage

# Konversi array menjadi DataFrame
data = pd.DataFrame(features, columns=['Bilirubin', 'Cholesterol'])
data['Stage'] = target

# Pisahkan data menjadi data latih (training) dan data uji (testing)
X_train, X_test, y_train, y_test = train_test_split(data[['Bilirubin', 'Cholesterol']], data['Stage'], test_size=0.2, random_state=42)

# Inisialisasi model regresi linear
model = LinearRegression()

# Melatih model dengan data latih
model.fit(X_train, y_train)

# Melakukan prediksi pada data uji
predictions = model.predict(X_test)

# Menghitung Mean Squared Error (MSE) sebagai metrik evaluasi
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test, predictions)
print(f'R-squared: {r2}')
```
